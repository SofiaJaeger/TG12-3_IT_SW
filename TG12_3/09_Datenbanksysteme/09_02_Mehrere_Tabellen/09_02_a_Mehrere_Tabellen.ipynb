### Datenbank Projektarbeiten
An der Turing-Gewerbeschule werden in der Jahrgangsstufe 1 Projekte angeboten. Dazu wurde die folgende Datenbank erstellt und mit den ersten Daten gefüllt:


```python
CREATE TABLE projekt (
    projekt_id INTEGER PRIMARY KEY,
    titel VARCHAR(100),
    fachbereich VARCHAR(50)
);

CREATE TABLE schueler (
    schueler_id INTEGER PRIMARY KEY,
    vorname VARCHAR(50),
    nachname VARCHAR(50),
    klasse VARCHAR(10),
    projekt_id INTEGER
);

INSERT INTO projekt VALUES
(10, 'Smart Home Steuerung', 'IoT'),
(11, 'Bibliotheksdatenbank', 'Datenbanken'),
(12, 'Robotik-Parcours', 'Robotik'),
(13, 'Wetterstation', 'Sensorik');

INSERT INTO schueler VALUES
(1, 'Mia', 'Keller', 'J1TG', 10),
(2, 'Jonas', 'Weber', 'J1TG', 10),
(3, 'Lina', 'Schmidt', 'J1TG', 11),
(4, 'Emir', 'Yilmaz', 'J1TG', 12),
(5, 'Noah', 'Fischer', 'J1TG', NULL);
```

#### Aufgabe 1: INNER JOIN

<u>__Code:__</u>

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
INNER JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

| Verbindungsart | Zweck |
| --- | --- |
| INNER JOIN | Zeigt nur Zeilen, bei denen der Wert aus schueler.projekt_id zu einem Wert aus projekt.projekt_id passt. |

#### Aufgabe 5: Verbindung mit WHERE ohne JOIN

<u>__Code:__</u>

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler, projekt
WHERE schueler.projekt_id = projekt.projekt_id;
```

<u>__Ausgabe:__</u>

|vorname|nachname|titel|fachbereich|
|---|---|---|---|
|Mia|Keller|Smart Home Steuerung|IoT|
|Jonas|Weber|Smart Home Steuerung|IoT|
|Lina|Schmidt|Bibliotheksdatenbank|Datenbanken|
|Emir|Yilmaz|Robotik-Parcours|Robotik|

<u>__Antworten:__</u>

1. Mia Keller, Jonas Weber, Lina Schmidt, Emir Yilmaz
2. Smart Home Steuerung, Bibliotheksdatenbank, Robotik-Parcours
3. Die Zeile mit Noah und die Zeile mit Wetterstation
4. Nirgends
5. Das Ergebnis entspricht dem INNer JOIN. Es erscheinen nur passende Kombinationen.

<u>__Beobachtung:__</u>

| Verbindungsart | Zweck |
| --- | --- |
| INNER JOIN | Verbindet zwei Tabellen über eine Bedingung in der WHERE-Klausel. Entspricht hier einem inneren Join. |
