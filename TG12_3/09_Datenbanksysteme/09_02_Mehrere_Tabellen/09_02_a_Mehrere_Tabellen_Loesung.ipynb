An der Turing-Gewerbeschule werden in der Jahrgangsstufe 1 Projekte angeboten. Dazu wurde die folgende Datenbank erstellt und mit den ersten Daten gefüllt:


```python
CREATE TABLE projekt (
    projekt_id INTEGER PRIMARY KEY,
    titel VARCHAR(100),
    fachbereich VARCHAR(50)
);

CREATE TABLE schueler (
    schueler_id INTEGER PRIMARY KEY,
    vorname VARCHAR(50),
    nachname VARCHAR(50),
    klasse VARCHAR(10),
    projekt_id INTEGER
);

INSERT INTO projekt VALUES
(10, 'Smart Home Steuerung', 'IoT'),
(11, 'Bibliotheksdatenbank', 'Datenbanken'),
(12, 'Robotik-Parcours', 'Robotik'),
(13, 'Wetterstation', 'Sensorik');

INSERT INTO schueler VALUES
(1, 'Mia', 'Keller', 'J1TG', 10),
(2, 'Jonas', 'Weber', 'J1TG', 10),
(3, 'Lina', 'Schmidt', 'J1TG', 11),
(4, 'Emir', 'Yilmaz', 'J1TG', 12),
(5, 'Noah', 'Fischer', 'J1TG', NULL);
```

#### Aufgabe 1: INNER JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
INNER JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Musterlösung

**Ausgabe**

| vorname | nachname | titel | fachbereich |
|:--------|:---------|:------|:------------|
| Mia | Keller | Smart Home Steuerung | IoT |
| Jonas | Weber | Smart Home Steuerung | IoT |
| Lina | Schmidt | Bibliotheksdatenbank | Datenbanken |
| Emir | Yilmaz | Robotik-Parcours | Robotik |

Noah fehlt, weil seine projekt_id den Wert NULL hat. Die Wetterstation fehlt, weil kein Schüler diesem Projekt zugeordnet ist.

| Verbindungsart | Zweck |
| --- | --- |
| INNER JOIN | Zeigt nur Zeilen, bei denen der Wert aus schueler.projekt_id zu einem Wert aus projekt.projekt_id passt. |

#### Aufgabe 2: LEFT JOIN


```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
LEFT JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Musterlösung

**Ausgabe**

| vorname | nachname | titel | fachbereich |
|:--------|:---------|:------|:------------|
| Mia | Keller | Smart Home Steuerung | IoT |
| Jonas | Weber | Smart Home Steuerung | IoT |
| Lina | Schmidt | Bibliotheksdatenbank | Datenbanken |
| Emir | Yilmaz | Robotik-Parcours | Robotik |
| Noah | Fischer | *null* | *null* |


Alle Zeilen aus der linken Tabelle schueler erscheinen. Wenn kein passendes Projekt gefunden wird, werden die Spalten aus projekt mit NULL gefüllt.

| Verbindungsart | Zweck |
| --- | --- |
| LEFT JOIN | Zeigt alle Zeilen der linken Tabelle und ergänzt passende Daten aus der rechten Tabelle. |


#### Aufgabe 3: RIGHT JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
RIGHT JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Musterlösung

**Ausgabe**

| vorname | nachname | titel | fachbereich |
|:--------|:---------|:------|:------------|
| Mia | Keller | Smart Home Steuerung | IoT |
| Jonas | Weber | Smart Home Steuerung | IoT |
| Lina | Schmidt | Bibliotheksdatenbank | Datenbanken |
| Emir | Yilmaz | Robotik-Parcours | Robotik |
| *null* | *null* | Wetterstation | Sensorik |


Alle Zeilen aus der rechten Tabelle projekt erscheinen. Die Wetterstation erscheint, obwohl kein Schüler zugeordnet ist. Die Schüler-Spalten enthalten dann NULL.

| Verbindungsart | Zweck |
| --- | --- |
| RIGHT JOIN | Zeigt alle Zeilen der rechten Tabelle und ergänzt passende Daten aus der linken Tabelle. |


#### Aufgabe 4: FULL OUTER JOIN


```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler
FULL OUTER JOIN projekt
ON schueler.projekt_id = projekt.projekt_id;
```

#### Musterlösung

**Ausgabe**

| vorname | nachname | titel | fachbereich |
|:--------|:---------|:------|:------------|
| Mia | Keller | Smart Home Steuerung | IoT |
| Jonas | Weber | Smart Home Steuerung | IoT |
| Lina | Schmidt | Bibliotheksdatenbank | Datenbanken |
| Emir | Yilmaz | Robotik-Parcours | Robotik |
| Noah | Fischer | *null* | *null* |
| *null* | *null* | Wetterstation | Sensorik |

Alle Zeilen aus beiden Tabellen erscheinen. Noah erscheint ohne Projekt. Die Wetterstation erscheint ohne zugeordneten Schüler.

| Verbindungsart | Zweck |
| --- | --- |
| FULL OUTER JOIN | Zeigt alle Zeilen beider Tabellen. Nicht passende Zeilen werden mit NULL ergänzt. |

#### Aufgabe 5: Verbindung mit WHERE ohne JOIN

```python
SELECT
    schueler.vorname,
    schueler.nachname,
    projekt.titel,
    projekt.fachbereich
FROM schueler, projekt
WHERE schueler.projekt_id = projekt.projekt_id;
```

#### Musterlösung

**Ausgabe**

| vorname | nachname | titel | fachbereich |
|:--------|:---------|:------|:------------|
| Mia | Keller | Smart Home Steuerung | IoT |
| Jonas | Weber | Smart Home Steuerung | IoT |
| Lina | Schmidt | Bibliotheksdatenbank | Datenbanken |
| Emir | Yilmaz | Robotik-Parcours | Robotik |


Das Ergebnis entspricht dem INNER JOIN. Es erscheinen nur passende Kombinationen.

| Verbindungsart | Zweck |
| --- | --- |
| Verbindung mit WHERE | verbindet zwei Tabellen über eine Bedingung in der WHERE-Klausel. Entspricht hier einem inneren Join. |

### Ergebnissicherung
| Verbindungsart | Was wird angezeigt? | Zweck | Beispiel |
| --- | --- | --- | --- |
| INNER JOIN | Nur passende Zeilen aus beiden Tabellen | Vollständige Zuordnungen anzeigen | ![INNER JOIN](Bilder/INNER-JOIN.png) |
| LEFT JOIN | Alle Zeilen der linken Tabelle | Fehlende Zuordnungen links erkennen | ![LEFT JOIN](Bilder/LEFT-JOIN.png) |
| RIGHT JOIN | Alle Zeilen der rechten Tabelle | Nicht genutzte Einträge rechts erkennen | ![RIGHT JOIN](Bilder/RIGHT-JOIN.png) |
| FULL OUTER JOIN | Alle Zeilen aus beiden Tabellen | Alle Zuordnungen und alle Lücken sichtbar machen | ![OUTER JOIN](Bilder/OUTER-JOIN.png) |
| WHERE ohne JOIN | Nur passende Zeilen | Ältere Schreibweise für einen inneren Join | ![WHERE](Bilder/WHERE.png) |
