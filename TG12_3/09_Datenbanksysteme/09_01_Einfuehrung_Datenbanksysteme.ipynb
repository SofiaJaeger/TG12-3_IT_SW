#### **🧩 2. Modellierung verstehen**

2.1 **Beschreibe** das Beispiel „Mitarbeiter – Projekt“.   
- Ein Mitarbeiter hat die Eigenschaften Vorname, Nachname und Alter. Ein Projekt hat die Eigenschaften Name, Kunde und Budget.  
- Der Mitarbeiter und das Projekt haben bestimmte Beziehungen.  
- Ein Mitarbeiter arbeitet höchstens an einem Projekt. An einem Projekt können mehrere Mitarbeiter arbeiten.  

2.2 **Bestimme** die Kardinalität der Beziehung und **begründe** deine Entscheidung.  
- Ein Mitarbeiter kann höchstens in einem Projekt arbeitern. Die Kardinalität ist 1.  
- Ein Projekt kann mehrere Mitarbeiter haben. Die Kardinalität ist n.  

2.3 **Erkläre**, warum man Daten zuerst modelliert und nicht direkt programmiert.  
- Weil es die Grundlage ist, und damit man die Beziehungen der Objekte besser verstehen.
--- 

#### **🏗️ 3. Relationales Modell**

3.1 **Erläutere**, wie aus einem ER-Diagramm Tabellen entstehen.
- Man erstellt eine Tabelle mit dem Namen des Objkektes, und schreibt die Eigenschaften hinein.
- Man schreibt in SQL, wie man eine Tabelle hinzufügt (CREATE TABLE) und übergibt die Eigenschaften.

3.2 **Beschreibe** die Bedeutung von:

- Primärschlüssel:
- Ein Primärschlüssen ist, wenn ein Attribut ganz eindeutig zugeordnet identifiziert ist.  

- Fremdschlüssel:
- Ein Fremdschlüssen ist, wenn ein Attribut auf ein anderes Schlüsselattribut in einer anderen Tabelle zeigt.  

3.3 **Erkläre**, warum Beziehungen in Tabellen dargestellt werden müssen.  
- Mann muss Beziehungen darstellen, um zu wissen, was bei Änderungen passiert.
---

#### **⚙️ 4. Integrität und Datenqualität**

4.1 **Nenne** Maßnahmen zur Sicherstellung der Datenintegrität.  
- Man muss Datentypen nennen.  
- Man kann Rechte vergeben, also wer was auf welcher Tabelle ausführen darf.  
- Man kann bestimmen welche Transaktionen man auf den Tabellen durchführen kann, also wie wir Daten einfügen oder ändern und was dann passiert.  


4.2 **Erkläre**, warum fehlerhafte Daten problematisch sind.
- Weil dann unklar ist, was mit den Werten passiert.


4.3 **Gib** ein Beispiel für eine Verletzung der Integrität.  
- Wenn beispielsweise ein Projekt gelöscht wird, indem noch Mitarbeiter arbeiten, ist unklar was mit den Werten passiert.
---

#### **🧮 5. SQL und Anfragen**

5.1 **Erkläre**, was „deklarative Anfrage“ bedeutet.    
- Man programmiert nicht "Wie erzeuge ich ein Anfrageergebnis?", so wie in anderen Programmiersprachen, sondern man programmiert "Was soll im Anfrageergebnis stehen?".

5.2 **Beschreibe** den Unterschied zwischen:

- natürlicher Sprache:  
- Die natürliche Sprache ist, wie wir Menschen immer in Worten sprechen.  

- relationaler Algebra:  
- Die relationale Algebra ist in mathematischen und griechischen Buchstaben ausgedrückt (Theoretische Ebene).  

- SQL:  
- SQL ist der formulierte Programmcode, welcher ausdrückt, was man sehen möchte (Praktische Ebene).  
  
5.3 **Formuliere** eine einfache SQL-Anfrage in eigenen Worten (ohne Syntax).  
- Zeige alle Mitarbeiter die in einem bestimmten Projekt arbeiten.
