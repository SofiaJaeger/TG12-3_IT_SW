```python
def findeLagerplatz(self, pArtikelname:str) -> int:

    # Datentypzuweisung
    idxGefundenerLp: int = -1
    idx: int
    name: str
    frei: bool
    gefunden: bool = False

    # Lagerplatz suchen, in dem schon der selbe Artikel eingelagert ist
    for idx in range(0, self.aAnz-1, 1):
        name = self.derLagerplatz[idx].gibArtikelname()
        if(name == pArtikelname):
            gefunden = True
            idxGefundenerLp = idx
            self.dieGui.melden(f"Lagerplatz {idxGefundenerLp} gefunden")
    

    # Lagerplatz suchen, der noch leer ist
    if(gefunden == False):
        for idx in range(0, self.aAnz-1, 1):
            frei = self.derLagerplatz[idx].istFrei()
            if(frei == True):
                gefunden = True
                idxGefundenerLp = idx
                self.dieGui.melden("Ware wird eingelagert")
    
    # Lagerplatz ist voll
    if(gefunden == False):
        self.dieGui.melden("Lager voll, einlagern nicht möglich")
    
    return idxGefundenerLp
```
