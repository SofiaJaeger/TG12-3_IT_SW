### 9.1 requests.post()

- Es braucht die Imports request und json.  
- Es muss die URL mit Route angegeben werden.  
- Es muss ein Objekt zum Beispiel Auto oder Spieler = {"mit : Eigenschaften"} erschaffen werden.  
- Dem Response muss request mit dem Aufruf von .post (URL, und die json Zuweisen von Auto/Spieler) zugewiesen werden.  
- Der Statuscode (response, status_code) kann ausgegeben werden.  
- Dazu braucht es die if-else abfrage der Fehler:  
- Ist response mit dem Aufruf von status_code 200 oder 201, wurde das Objekt erfolgreich erstellt.  
- Ansonsten gibt es einen Fehler bei der Erstellung.  
- Beides mal wird von json der Aufruf .dumps gemacht, in welchem (response mit dem Aufruf von .json(), der Angabe eines indent, welcher 4 ist, und dem ensure_ascii, welcher falsch ist) aufgerufen wird.


| Schritt | Aufgabe | Werkzeug |
|----------|----------|-----------|
| 1 | Server empfängt Daten | `Flask` |
| 2 | Validierung der Daten | `Pydantic` |
| 3 | Rückgabe bei Erfolg oder Fehler | HTTP 201 / 400 |
| 4 | Client sendet Daten | `requests.post()` |
| 5 | Ergebnis anzeigen | `response.json()` |

So entsteht ein vollständiger Client-Server-Datenaustausch mit Pydantic-Validierung.

# 3. Model

### 8.2 Attributwerte überprüfen

Mit `Field()` können in Pydantic Regeln für einzelne Attribute festgelegt werden.  

- Es braucht von pydantic die Importe BaseModel, Field, field_validator, ValidationError.  
- Es wird eine Klasse erstellt, welcher (BaseModel) übergeben wird.  
- Eigenschaften der Objekte: werden mit Datentyp = Field (Daten auf Erlaubtheit geprüft).  
- Diese können sein: siehe Tabelle.  
- Dann muss die Live-Validierung bei späteren Änderungen aktiviert werden.  
- Das geschieht, indem model_config = {"das validate_assignment": welches Wahr ist} zugewiesen wird.  


| Option | Bedeutung | Beispiel | Wirkung |
|:--------|:-----------|:----------|:---------|
| `default` | Standardwert für das Feld | `Field(default=0)` | Wenn kein Wert übergeben wird, wird 0 gesetzt |
| `ge` | „greater or equal“ – Minimalwert | `Field(..., ge=0)` | Wert muss ≥ 0 sein |
| `gt` | „greater than“ – strenger Minimalwert | `Field(..., gt=0)` | Wert muss > 0 sein |
| `le` | „less or equal“ – Maximalwert | `Field(..., le=100)` | Wert muss ≤ 100 sein |
| `lt` | „less than“ – strenger Maximalwert | `Field(..., lt=100)` | Wert muss < 100 sein |
| `min_length` | Minimale Länge von Strings oder Listen | `Field(..., min_length=3)` | Prüft, dass Text oder Liste mindestens 3 Elemente hat |
| `max_length` | Maximale Länge von Strings oder Listen | `Field(..., max_length=20)` | Prüft, dass Text oder Liste höchstens 20 Elemente hat |
