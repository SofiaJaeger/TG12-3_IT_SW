### 1.8 Klassenattribte
Klassenattribute gehören nicht zu einem einzelnen Objekt, sondern zur Klasse selbst.  
Sie werden von allen Instanzen gemeinsam genutzt.

```python
class Schueler:
    anzahl = 0   # Klassenattribut

    def __init__(self, name):
        self.name = name
        Schueler.anzahl += 1
```

### 1.9 Klassenmethoden
Klassenmethoden werden mit dem Dekorator @classmethod gekennzeichnet.  
Sie erhalten als ersten Parameter die Klasse selbst (cls).  
Sie werden genutzt, um auf Klassenattribute zuzugreifen.  

```python
@classmethod
def statistik(cls):
    return f"Es gibt aktuell {cls.anzahl} Schüler."
```

### 3.2 Localhost

Der Begriff „localhost“ bezeichnet den eigenen Rechner, also das lokale System, auf dem ein Programm gerade ausgeführt wird.  

Technisch ist „localhost“ fest mit der IPv4-Adresse 127.0.0.1 verknüpft.  
Diese Adresse ist reserviert und verweist stets auf das lokale Gerät selbst, unabhängig von dessen tatsächlicher Netzwerkverbindung.  

Ein häufiges Einsatzgebiet von „localhost“ ist die Entwicklung und das Testen von Server-Anwendungen:  
Entwickler können einen lokalen Webserver starten und über http://localhost:12345 im Browser auf diesen zugreifen.

# 7 Flask

### 7.1 Request

In Flask ist `request` keine normale Variable, sondern ein “Kontext-gebundenes Proxy-Objekt”.  
Die Methode, die per `route` die Anfrage erhält, kann über `request` auf die Daten der Anfrage zugreifen:

| Attribut           | Beschreibung |
|--------------------|---------------|
| `request.method`   | Gibt die HTTP-Methode der Anfrage zurück, z. B. `"GET"`, `"POST"`, `"PUT"`, `"DELETE"`. |
| `request.args`     | Enthält die URL-Parameter (Query-Parameter) als Dictionary, z. B. bei `/page?name=Max`. |
| `request.form`     | Beinhaltet Formulardaten, die per `POST` mit `application/x-www-form-urlencoded` gesendet wurden. |
| `request.json`     | Liefert ein Dictionary der JSON-Daten im Request-Body, wenn der Client JSON sendet. |
| `request.data`     | Gibt den rohen Body der Anfrage als Byte-String zurück (z. B. wenn kein JSON-Format). |
| `request.headers`  | Ermöglicht Zugriff auf alle HTTP-Header der Anfrage. |
| `request.cookies`  | Beinhaltet alle gesendeten Cookies als Dictionary. |
| `request.path`     | Zeigt den angefragten Pfad der URL, z. B. `"/message"`. |
| `request.url`      | Gibt die vollständige URL der Anfrage zurück. |
| `request.remote_addr` | Enthält die IP-Adresse des Clients, der die Anfrage gesendet hat. |

# 8 Model/ Objektübertragung

### 8.1 Pydantic

Pydantic kann Objekte einfach in JSON umwandeln. (`model_dump()`, `model_dump_json()`)

```python
from pydantic import BaseModel

class Student(BaseModel):
    name: str
    age: int
    course: str

# Beispielobjekt
s = Student(name="Anna", age=21, course="Informatik")
print(s)
print(s.model_dump())  # Als Dictionary
```

### 8.2 Attributwerte überprüfen

Mit `Field()` können in Pydantic Regeln für einzelne Attribute festgelegt werden.  


| Option | Bedeutung | Beispiel | Wirkung |
|:--------|:-----------|:----------|:---------|
| `default` | Standardwert für das Feld | `Field(default=0)` | Wenn kein Wert übergeben wird, wird 0 gesetzt |
| `ge` | „greater or equal“ – Minimalwert | `Field(..., ge=0)` | Wert muss ≥ 0 sein |
| `gt` | „greater than“ – strenger Minimalwert | `Field(..., gt=0)` | Wert muss > 0 sein |
| `le` | „less or equal“ – Maximalwert | `Field(..., le=100)` | Wert muss ≤ 100 sein |
| `lt` | „less than“ – strenger Maximalwert | `Field(..., lt=100)` | Wert muss < 100 sein |
| `min_length` | Minimale Länge von Strings oder Listen | `Field(..., min_length=3)` | Prüft, dass Text oder Liste mindestens 3 Elemente hat |
| `max_length` | Maximale Länge von Strings oder Listen | `Field(..., max_length=20)` | Prüft, dass Text oder Liste höchstens 20 Elemente hat |

```python

from pydantic import BaseModel, Field, field_validator, ValidationError

class Auto(BaseModel):

    ps: int = Field(default=75, ge=0, le=500)
    verbrauch: float = Field(default=5.0)

    # Aktiviert Live-Validierung bei späteren Änderungen
    model_config = {"validate_assignment": True}
```

### 8.3 Falsche Attributwerte

```python

# Objekt erzeugen (gültige Werte)
a = Auto(ps=120, verbrauch=6.5)
print("Neues Auto:", a.model_dump())

# Änderung eines gültigen Wertes
a.ps = 90
print("Nach Änderung:", a.model_dump())

# Ungültiger Wert für PS
try:
    a.ps = -50
    print(a.model_dump())
except ValueError as e:
    print("Fehler:", e)
```

### 8.4 Operator **
Der Operator ** entpackt ein Dictionary in einzelne Schlüssel-Wert-Paare und übergibt sie als benannte Argumente an die Funktion oder den Konstruktor.

```python
from flask import request

@app.route("/student", methods=["POST"])
def receive_student():
    data = request.get_json()
    student = Student(**data)  # JSON → Objekt
    print("Empfangen:", student)
    return jsonify({"message": "Student empfangen", "student": student.model_dump()})
```

# 9. Client: Daten an Server senden

### 9.1 requests.post()

```python
import request
import json

URL = "http://127.0.0.1:12345/auto"

auto_daten = {
    "marke" : "Audi",
    "ps" : 150,
}

response = request.post(URL, json=auto_daten)
print("Statuscode:", response,status_code)

if response.status_code == 200 or response.status_code == 201:
    print("Auto erfolgreich erstellt:")
    print(json.dumps(response.json(), indent=4, ensure_ascii=False))
else:
    print("Fehler bei der Erstellung:")
    print(json.dumps(response.json(), indent=4, ensure_ascii=False))
```

### 9.2 Übersicht

| Schritt | Aufgabe | Werkzeug |
|----------|----------|-----------|
| 1 | Server empfängt Daten | `Flask` |
| 2 | Validierung der Daten | `Pydantic` |
| 3 | Rückgabe bei Erfolg oder Fehler | HTTP 201 / 400 |
| 4 | Client sendet Daten | `requests.post()` |
| 5 | Ergebnis anzeigen | `response.json()` |

So entsteht ein vollständiger Client-Server-Datenaustausch mit Pydantic-Validierung.

# 11. CORS

### 11.1 Erklärung
CORS ist ein Sicherheitsmechanismus im Browser, der verhindert, dass eine Webseite einfach Daten von einem anderen Server lädt.


### 11.2 Mini-Webserver starten
Du kannst einen kleinen Server starten, der die Dateien korrekt über HTTP ausliefert:

```bash
cd pfad/zum/ordner
python -m http.server 12345
```

Danach erreichst du deine Seite über:

```
http://127.0.0.1:12345
```

Jetzt sendet der Browser den Header:

```
Origin: http://127.0.0.1:12345
```

Flask kann diese Origin erlauben, und CORS funktioniert korrekt.

### 11.3 Problemlösung in Flask

Für reale Projekte solltest du die erlaubten Origins einschränken:

```python
from flask import Flask, jsonify
from flask_cors import CORS

CORS(app)

#--------ODER---------

app = Flask(__name__)
CORS(app, origins=[
    "http://127.0.0.1:12345",
    "http://localhost:12345"
])

@app.get("/api/test")
def test():
    return jsonify(ok=True, msg="Nur erlaubte Origins dürfen zugreifen.")

app.run(port=12345, debug=True)
```

So dürfen nur Anfragen von diesen Adressen auf die API zugreifen.
