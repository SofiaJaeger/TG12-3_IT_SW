### 1.8 Klassenattribte
Klassenattribute gehören nicht zu einem einzelnen Objekt, sondern zur Klasse selbst.  
Sie werden von allen Instanzen gemeinsam genutzt.  
Einsatzgebiete sind Zähler, Konstanten oder allgemein Daten, die für alle Objekte einer Klasse gleich gelten sollen.  

```python
class Schueler:
    anzahl = 0   # Klassenattribut

    def __init__(self, name):
        self.name = name
        Schueler.anzahl += 1
```

### 1.9 Klassenmethoden
Klassenmethoden werden mit dem Dekorator @classmethod gekennzeichnet.  
Sie erhalten als ersten Parameter die Klasse selbst (cls).  
Sie werden genutzt, um auf Klassenattribute zuzugreifen.  

```python
@classmethod
def statistik(cls):
    return f"Es gibt aktuell {cls.anzahl} Schüler."
```

### 3.2 Localhost

Der Begriff „localhost“ bezeichnet den eigenen Rechner, also das lokale System, auf dem ein Programm gerade ausgeführt wird.  

Technisch ist „localhost“ fest mit der IPv4-Adresse 127.0.0.1 verknüpft.  
Diese Adresse ist reserviert und verweist stets auf das lokale Gerät selbst, unabhängig von dessen tatsächlicher Netzwerkverbindung.  

Ein häufiges Einsatzgebiet von „localhost“ ist die Entwicklung und das Testen von Server-Anwendungen:  
Entwickler können einen lokalen Webserver starten und über http://localhost:12345 im Browser auf diesen zugreifen.
