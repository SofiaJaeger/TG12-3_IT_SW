### 1.8 Klassenattribte
Klassenattribute gehören nicht zu einem einzelnen Objekt, sondern zur Klasse selbst.  
Sie werden von allen Instanzen gemeinsam genutzt.  
Einsatzgebiete sind Zähler, Konstanten oder allgemein Daten, die für alle Objekte einer Klasse gleich gelten sollen.  

```python
class Schueler:
    anzahl = 0   # Klassenattribut

    def __init__(self, name):
        self.name = name
        Schueler.anzahl += 1
```

### 1.9 Klassenmethoden
Klassenmethoden werden mit dem Dekorator @classmethod gekennzeichnet.  
Sie erhalten als ersten Parameter die Klasse selbst (cls).  
Sie werden genutzt, um auf Klassenattribute zuzugreifen.  

```python
@classmethod
def statistik(cls):
    return f"Es gibt aktuell {cls.anzahl} Schüler."
```

### 3.2 Localhost

Der Begriff „localhost“ bezeichnet den eigenen Rechner, also das lokale System, auf dem ein Programm gerade ausgeführt wird.  

Technisch ist „localhost“ fest mit der IPv4-Adresse 127.0.0.1 verknüpft.  
Diese Adresse ist reserviert und verweist stets auf das lokale Gerät selbst, unabhängig von dessen tatsächlicher Netzwerkverbindung.  

Ein häufiges Einsatzgebiet von „localhost“ ist die Entwicklung und das Testen von Server-Anwendungen:  
Entwickler können einen lokalen Webserver starten und über http://localhost:12345 im Browser auf diesen zugreifen.

### 8.2 Attributwerte überprüfen

Mit `Field()` können in Pydantic Regeln für einzelne Attribute festgelegt werden.  


| Option | Bedeutung | Beispiel | Wirkung |
|:--------|:-----------|:----------|:---------|
| `default` | Standardwert für das Feld | `Field(default=0)` | Wenn kein Wert übergeben wird, wird 0 gesetzt |
| `ge` | „greater or equal“ – Minimalwert | `Field(..., ge=0)` | Wert muss ≥ 0 sein |
| `gt` | „greater than“ – strenger Minimalwert | `Field(..., gt=0)` | Wert muss > 0 sein |
| `le` | „less or equal“ – Maximalwert | `Field(..., le=100)` | Wert muss ≤ 100 sein |
| `lt` | „less than“ – strenger Maximalwert | `Field(..., lt=100)` | Wert muss < 100 sein |
| `min_length` | Minimale Länge von Strings oder Listen | `Field(..., min_length=3)` | Prüft, dass Text oder Liste mindestens 3 Elemente hat |
| `max_length` | Maximale Länge von Strings oder Listen | `Field(..., max_length=20)` | Prüft, dass Text oder Liste höchstens 20 Elemente hat |

```python

from pydantic import BaseModel, Field, field_validator, ValidationError

class Auto(BaseModel):

    ps: int = Field(default=75, ge=0, le=500)
    verbrauch: float = Field(default=5.0)

    # Aktiviert Live-Validierung bei späteren Änderungen
    model_config = {"validate_assignment": True}
```

### 8.3 Falsche Attributwerte

```python

# Objekt erzeugen (gültige Werte)
a = Auto(ps=120, verbrauch=6.5)
print("Neues Auto:", a.model_dump())

# Änderung eines gültigen Wertes
a.ps = 90
print("Nach Änderung:", a.model_dump())

# Ungültiger Wert für PS
try:
    a.ps = -50
    print(a.model_dump())
except ValueError as e:
    print("Fehler:", e)
```

### 8.4 Operator **
Der Operator `**` entpackt ein Dictionary in einzelne Schlüssel-Wert-Paare und übergibt sie als benannte Argumente an die Funktion oder den Konstruktor.

```python
from flask import request

@app.route("/student", methods=["POST"])
def receive_student():
    data = request.get_json()
    student = Student(**data)  # JSON → Objekt
    print("Empfangen:", student)
    return jsonify({"message": "Student empfangen", "student": student.model_dump()})
```
